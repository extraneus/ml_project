```python
import pandas as pd
import numpy as np
from sklearn import tree
import pydotplus
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

# Seperate Feature Columns

```python
x = df.drop(['DEATH_EVENT'], axis = 1)
x
```

# Seperate Class Column

```python
y = df['DEATH_EVENT']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
def evaluate_model(model, model_name):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f"\n{model_name} Results:")
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
dt_model = tree.DecisionTreeClassifier()
evaluate_model(dt_model, "Decision Tree")
```

```python
rf_model = RandomForestClassifier(random_state=42)
evaluate_model(rf_model, "Random Forest")
```

```python
lr_model = LogisticRegression(random_state=42, max_iter=1000) # Increased max_iter
evaluate_model(lr_model, "Logistic Regression")
```

# ANN

```python
from tensorflow.keras.utils import to_categorical

dummy_y = to_categorical(y)
```

```python
dummy_y.shape
```

```python
dummy_y
```

# Train, Test & Split our dataset into 70:30

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,dummy_y,test_size=.3,random_state=1)
```

# Create an ANN model with Kears, Fit our Train dtata into our model

```python
model = Sequential()
model.add(Dense(12, input_shape=(12,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(2, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(xtrain, ytrain, batch_size = 5, epochs = 200)
```

```python
_, accuracy = model.evaluate(xtest, ytest)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
y_pred = model.predict(xtest)
y_pred=np.argmax(y_pred, axis=1)
```

```python
y_acctual=np.argmax(ytest, axis=1)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_acctual, y_pred)
print(cm)
```

```python
print(classification_report(y_acctual, y_pred))
```
