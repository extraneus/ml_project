```python
from collections import deque


adj_list = {
    'A': ['B', 'D'],
    'B': ['A', 'C'],
    'C': ['B'],
    'D': ['A', 'E', 'F'],
    'E': ['D', 'F', 'G'],
    'F': ['D', 'E', 'H'],
    'G': ['E', 'H'],
    'H': ['F', 'G']
}

visited = {}
level = {}
parent = {}
bfs_traversal_output = []

for node in adj_list:
    visited[node] = False
    parent[node] = None
    level[node] = -1

que = deque()
start_node = 'A'
visited[start_node] = True
level[start_node] = 0
que.append(start_node)
while que:
    u = que.popleft()
    bfs_traversal_output.append(u)

    for neighbor in adj_list[u]:
        if not visited[neighbor]:
            visited[neighbor] = True
            parent[neighbor] = u
            level[neighbor] = level[u] + 1
            que.append(neighbor)

print("BFS Traversal Output:", bfs_traversal_output)

```

```python
def bfs(graph, start_node, end_node):

    solution = []
    costs = 0

    frontier = []
    visited = []

    frontier.append(start_node)
    visited.append(start_node)

    while frontier:

        selected_node = frontier.pop(0)

        if selected_node == end_node:
            solution.append(selected_node)
            break

        solution.append(selected_node)

        for neighbour in graph[selected_node]:
            if neighbour not in visited:
                frontier.append(neighbour)
                visited.append(neighbour)

        costs += 1

    return solution, costs
```

```python
from collections import deque

def bfs_8_puzzle(initial, goal):
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    initial = tuple(tuple(row) for row in initial)
    goal = tuple(tuple(row) for row in goal)

    queue = deque([(initial, [])])
    visited = set()
    visited.add(initial)

    while queue:
        current_state, path = queue.popleft()
        if current_state == goal:
            return path
        for i in range(3):
            for j in range(3):
                if current_state[i][j] == 0:
                    blank_x, blank_y = i, j
                    break
        for dx, dy in directions:
            new_x, new_y = blank_x + dx, blank_y + dy
            if 0 <= new_x < 3 and 0 <= new_y < 3:
                new_state = [list(row) for row in current_state]
                new_state[blank_x][blank_y], new_state[new_x][new_y] = new_state[new_x][new_y], new_state[blank_x][blank_y]
                new_state = tuple(tuple(row) for row in new_state)

                if new_state not in visited:
                    visited.add(new_state)
                    queue.append((new_state, path + [(new_x, new_y)]))

    return None

initial = [
    [1, 2, 3],
    [4, 0, 5],
    [7, 8, 6]
]

goal = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

solution = bfs_8_puzzle(initial, goal)
if solution:
    print("Solution Found:")
    for step in solution:
        print(f"Move blank tile to position {step}")
else:
    print("No solution exists.")
```

```python
from collections import defaultdict
tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

def depth_limited_dfs(node, goal, depth, visited, path):

    if node == goal:
        path.append(node)
        return True
    if depth <= 0:
        return False
    visited.add(node)
    path.append(node)
    for neighbor in tree[node]:
        if neighbor not in visited:
            if depth_limited_dfs(neighbor, goal, depth - 1, visited, path):
                return True
    path.pop()
    return False

def iterative_deepening_dfs(start, goal):

    depth = 0
    while True:
        visited = set()
        path = []

        print(f"Searching with depth limit: {depth}")
        if depth_limited_dfs(start, goal, depth, visited, path):
            return path

        depth += 1

        if depth > 100:
            return None

start_node = 'A'
goal_node = 'G'

path = iterative_deepening_dfs(start_node, goal_node)
if path:
    print(f"Path found: {' -> '.join(path)}")
else:
    print("No path found.")
```

```python
from collections import defaultdict

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)

    def add_edge(self, u, v):
        self.graph[u].append(v)

    def is_cyclic_util(self, node, visited, rec_stack):
        visited[node] = True
        rec_stack[node] = True

        for neighbor in self.graph[node]:
            if not visited[neighbor]:
                if self.is_cyclic_util(neighbor, visited, rec_stack):
                    return True
            elif rec_stack[neighbor]:
                return True

        rec_stack[node] = False
        return False

    def is_cyclic(self):
        visited = {node: False for node in self.graph}
        rec_stack = {node: False for node in self.graph}
        for node in self.graph:
            if not visited[node]:
                if self.is_cyclic_util(node, visited, rec_stack):
                    return True

        return False

graph = Graph()
graph.add_edge(0, 1)
graph.add_edge(1, 2)
graph.add_edge(2, 0)

if graph.is_cyclic():
    print("The graph contains a cycle.")
else:
    print("The graph does not contain a cycle.")
```
